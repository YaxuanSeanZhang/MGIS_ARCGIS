Evaluation Metrics

Confusion Matrix

```python
from sklearn.metrics import confusion_matrix

# Create some sample labels and predictions
y_true = [1, 0, 1, 1, 0, 1, 0, 1, 0, 0]
y_pred = [1, 0, 0, 1, 0, 1, 1, 1, 1, 0]

# Compute the confusion matrix
cm = confusion_matrix(y_true, y_pred)

# Print the confusion matrix
print(cm)
```

Accuracy

```python
from sklearn.metrics import accuracy_score

# Compute the accuracy
acc = accuracy_score(y_true, y_pred)

# Print the accuracy
print("Accuracy:", acc)
```

Precision

```python
from sklearn.metrics import precision_score

# Compute the precision
precision = precision_score(y_true, y_pred)

# Print the precision
print("Precision:", precision)
```

Recall

```python
from sklearn.metrics import recall_score

# Compute the recall
recall = recall_score(y_true, y_pred)

# Print the recall
print("Recall:", recall)
```

True Positives

```python
conf_matrix = confusion_matrix(y_true, y_pred)

# Extract the number of true positives from the confusion matrix
true_positives = conf_matrix[1, 1]

# Print the number of true positives
print("True positives:", true_positives)
```

False Positives

```python
conf_matrix = confusion_matrix(y_true, y_pred)

# Extract the number of false positives from the confusion matrix
false_positives = conf_matrix[0, 1]

# Print the number of false positives
print("False positives:", false_positives)
```

Receiver Operator Characteristic (ROC) Curve and Area Under the Curve

```python
from sklearn.metrics import roc_curve, roc_auc_score
import matplotlib.pyplot as plt

# Create some sample labels and predictions
y_true = [1, 0, 1, 1, 0, 1, 0, 1, 0, 0]
y_pred = [0.9, 0.1, 0.8, 0.7, 0.3, 0.6, 0.2, 0.5, 0.4, 0.2]

# Compute the false positive rate, true positive rate, and classification thresholds
fpr, tpr, thresholds = roc_curve(y_true, y_pred)

# Compute the AUC
auc = roc_auc_score(y_true, y_pred)

# Plot the ROC curve
plt.plot(fpr, tpr, label=f"AUC = {auc:.2f}")
plt.plot([0, 1], [0, 1], "--", color="gray", label="Random")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("Receiver Operating Characteristic (ROC) Curve")
plt.legend()
plt.show()
```

R-squared

```python
from sklearn.metrics import r2_score
import numpy as np

# Create some sample data
X = np.array([[1, 2], [3, 4], [5, 6], [7, 8]])
y_true = np.array([3, 7, 11, 15])

# Fit a linear regression model
from sklearn.linear_model import LinearRegression
reg = LinearRegression().fit(X, y_true)

# Compute R-squared
y_pred = reg.predict(X)
r2 = r2_score(y_true, y_pred)

print(f"R-squared: {r2:.2f}")
```

Adjusted R-squared

```python
# Compute adjusted R-squared
n = X.shape[0]
p = X.shape[1]
adj_r2 = 1 - (1 - r2) * (n - 1) / (n - p - 1)

print(f"Adjusted R-squared: {adj_r2:.2f}")
```

Root Mean Square Error

```python
from sklearn.metrics import mean_squared_error
from math import sqrt

# Calculate RMSE
rmse = sqrt(mean_squared_error(y_true, y_pred))

print(f"RMSE: {rmse:.2f}")
```

Mean Absolute Error

```python
from sklearn.metrics import mean_absolute_error

# Calculate the MAE
mae = mean_absolute_error(y_true, y_pred)

print("MAE:", mae)
```

Residual Standard Error

```python
from sklearn.metrics import mean_squared_error

# Calculate the RSE
rse = np.sqrt(mean_squared_error(y_true, y_pred))

print("Residual Standard Error: ", rse)
```

Akaike’s Information Criterion (AIC)

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import statsmodels.api as sm

# Fit the linear regression model using scikit-learn
model = LinearRegression().fit(X, y_true)

# Calculate the mean squared error
residuals = y_true - model.predict(X)
sigma = np.std(residuals)

# Calculate the AIC
n = len(y_true)
k = model.coef_.shape[0] + 1  # Add 1 for the intercept term
log_likelihood = -n/2*np.log(2*np.pi*sigma**2) - 1/(2*sigma**2)*np.sum(residuals**2)

aic = 2 * k - 2 * log_likelihood

print("AIC: ", aic)
```

Bayesian Information Criterion (BIC)

```python
bic = -2*log_likelihood + k *np.log(n)

print('BIC =', bic)
```
