```python
import requests
import zipfile
from zipfile import ZipFile
import os
import arcpy
```

```python
def get_data(url, local_directory):
    response = requests.get(url)
    file_name = os.path.join(local_directory, url.split("/")[-1])

    with open(file_name, "wb") as f:
        f.write(response.content)
        
    with ZipFile(file_name, "r") as zip_ref:
        zip_ref.extractall(local_directory)
```

```python
url = r"https://resources.gisdata.mn.gov/pub/gdrs/data/pub/us_mn_state_dot/trans_roads_mndot_tis/shp_trans_roads_mndot_tis.zip"
local_directory = "E:\coursework\ARCGISI\Lab0\data"

# Create the local directory if it doesn't exist
if not os.path.exists(local_directory):
    os.makedirs(local_directory)

get_data(url,local_directory)   
```
